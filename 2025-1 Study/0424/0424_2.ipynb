## 결측치 처리
Arrival Delay in Minutes: 전체 데이터 수에 비해 극히 적은 결측치 가짐 & 이상치가 크기 때문에 중앙값(0)으로 대체

+ 추가적으로 test dataset이 별도로 주어지지 않은 데이터셋이므로 trainset의 20% 분할하여 test dataset으로 사용

```python
train['satisfaction'] = train['satisfaction'].astype('category').cat.codes
train.head()
```

```python
train['satisfaction'].value_counts()
```

```python
X = train.drop(columns='satisfaction')
y = train['satisfaction']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

```python
# 진짜 결측치 처리
median = X_train['Arrival Delay in Minutes'].median(0)

X_train['Arrival Delay in Minutes'] = X_train['Arrival Delay in Minutes'].fillna(median)
X_test['Arrival Delay in Minutes'] = X_test['Arrival Delay in Minutes'].fillna(median)
```

```python
X_train.isna().sum()
```

```python
for col in object_col[1:]:
    unique_list = X_train[col].unique()
    mapping = {}
    # train dataset 기준으로 매핑
    for u in range(len(unique_list)):
        mapping[unique_list[u]] = u
    
    X_train[col] = X_train[col].map(mapping)
    X_test[col] = X_test[col].map(mapping)
```

## 스케일링

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train['Flight Distance'] = scaler.fit_transform(pd.DataFrame(X_train['Flight Distance']))
```

```python
X_train['Flight Distance'] = scaler.fit_transform(pd.DataFrame(X_train['Flight Distance']))
```

```python
X_test['Flight Distance'] = scaler.transform(pd.DataFrame(X_test['Flight Distance']))
```

```python
X_train['Departure Delay in Minutes'] = scaler.fit_transform(pd.DataFrame(X_train['Departure Delay in Minutes']))
X_test['Departure Delay in Minutes'] = scaler.transform(pd.DataFrame(X_test['Departure Delay in Minutes']))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_validate, StratifiedKFold
from sklearn.metrics import accuracy_score
```

```python
models = {
    'RF': RandomForestClassifier(),
    'GD': GradientBoostingClassifier(),
    'LR': LogisticRegression()
}
```

```python
for name, model in models.items():
    score = cross_validate(model, X_train, y_train, scoring='accuracy', cv=StratifiedKFold(), n_jobs=-1, return_train_score=True)
    print(f"{name}: {np.mean(score['train_score'])}, {np.mean(score['test_score'])}")
```

```python
params = {
    'max_depth': range(2, 20, 2),
    'n_estimators': [100, 200, 300]
}
```

```python
grid = GridSearchCV(RandomForestClassifier(), param_grid=params, cv=5, n_jobs=-1, verbose=1, scoring='accuracy')
grid.fit(X_train, y_train)
```

```python
print(f'{grid.best_score_}, {grid.best_params_}') 
```

```python
model = grid.best_estimator_
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred
```

```python
accuracy_score(y_test, pred)
```
