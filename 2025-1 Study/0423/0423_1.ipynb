```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")
```

```python
import numpy as np
```

```python
train = pd.read_csv("open/train.csv")
test = pd.read_csv("open/test.csv")
```

```python
train.head()
```

```python
train.info()
```

```python
test.head()
```

```python
train.describe()
```

```python
test = test.drop(columns=['ID'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# id부터 타겟변수 전까지 scaling 진행
train_scaler = scaler.fit_transform(train[train.columns[1:-1]])
test_scaler = scaler.transform(test)
```

```python
features =  ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

for i, feature in enumerate(features):
    row = int(i / 4)
    col = i % 4

    sns.regplot(x=feature, y='MEDV', data=train, ax=axes[row][col])
```

```python
train = train.drop(columns='ID')
```

```python
corr_order = train.corr().loc[:'LSTAT', 'MEDV'].abs().sort_values(ascending=False)
corr_order
```

```python
sns.displot(x='MEDV', data=train, kind='hist')
```

```python
mask = np.triu(np.ones_like(train.corr(), dtype=bool))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
corr = train.corr()
fig, axes = plt.subplots(figsize=(10, 8))
sns.heatmap(corr, mask=mask, cmap=cmap, linewidths=.5, fmt='.1f', annot=True)
```

```python
X_train = train.iloc[:, :-1]
X_train
X_train.iloc[:, :] = train_scaler[:, :]
```

```python
X_train.head()
```

```python
y_train = train['MEDV']
```

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(test_scaler)
```

```python
from sklearn.metrics import mean_squared_error

train_pred = lr_model.predict(X_train)

train_mse = mean_squared_error(y_train, train_pred)
train_mse
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import Ridge, Lasso
```

```python
models = {
    'RandomForest': RandomForestRegressor(),
    'GradientBoosting': GradientBoostingRegressor(),
    'Ridge': Ridge(),
    'Lasso': Lasso()
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_train)
    mse = mean_squared_error(y_train, pred)
    print(f'{name} mse : {mse}')
```

```python
rf_model = RandomForestRegressor()
rf_model.fit(X_train, y_train)
```

```python
feature_importance = pd.Series(rf_model.feature_importances_, index=X_train.columns.values)
feature_importance
```

```python
feature_importance.sort_values(ascending=True).plot(kind='barh')
```

```python
X_train_cut = X_train.drop(columns=['ZN', 'CHAS'])
```

```python
rf_model_cut = RandomForestRegressor()
rf_model_cut.fit(X_train_cut, y_train)
```

```python
rf_model_cut_pred = rf_model_cut.predict(X_train_cut)
rf_model_cut_mse = mean_squared_error(y_train, rf_model_cut_pred)
rf_model_cut_mse
```

```python
help(RandomForestRegressor)
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'n_estimators': [100, 200, 500, 1000],
    'max_depth': [i for i in range(1, 10, 2)],
    'min_samples_split': [i for i in range(2, 10, 2)]
}
params
```

```python
grid = GridSearchCV(estimator=rf_model_cut, param_grid=params, cv=5)
grid = grid.fit(X_train, y_train)
print(f'best score parameters: {grid.best_score_:.4f}')

print("Score: ")
for key, value in grid.best_params_.items():
    print(f'\t {key}:{value}')
```

```python
rf_model_cv = RandomForestRegressor(max_depth=9, n_estimators=200, min_samples_split=2)
rf_model_cv.fit(X_train, y_train)
rf_model_cv_pred = rf_model_cv.predict(X_train)
rf_model_cv_mse = mean_squared_error(y_train, rf_model_cv_pred)
rf_model_cv_mse
```

```python
pred = rf_model_cv.predict(test_scaler)

submit = pd.read_csv('open/sample_submission.csv')
submit['MEDV'] = pred
submit.head()
```

```python
submit.to_csv("submit.csv", index=False)
```
