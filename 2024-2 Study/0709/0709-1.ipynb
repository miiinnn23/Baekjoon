```python
# 변수 별 분포가 다르므로, 표준화 작업 수행
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
train[numeric_col] = scaler.fit_transform(train[numeric_col])
```

```python
train.describe()
```

```python
train['type'] = train['type'].astype('category').cat.codes
```

```python
train.info()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, HistGradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

from sklearn.model_selection import KFold, cross_validate, cross_val_score, StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
X_train = train.drop(columns='quality')
Y_train = train['quality']
```

```python
dt = DecisionTreeClassifier()

score = cross_validate(dt, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
rf = RandomForestClassifier()

score = cross_validate(rf, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
gb = GradientBoostingClassifier()

score = cross_validate(gb, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
hgb = HistGradientBoostingClassifier()

score = cross_validate(hgb, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
svm = SVC()

score = cross_validate(svm, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
nb = GaussianNB()

score = cross_validate(nb, X_train, Y_train, return_train_score=True, cv=StratifiedKFold(), n_jobs=-1)
print(np.mean(score['train_score']), np.mean(score['test_score']))
```

```python
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': range(2, 20, 2)
}

grid_rf = GridSearchCV(RandomForestClassifier(), rf_params, cv=5, n_jobs=-1)
grid_rf.fit(X_train, Y_train)
pred = grid_rf.predict(X_train)

acc = accuracy_score(Y_train, pred)
```
