```python
# Visualization 연습

x = df.drop(columns='diagnosis')
y = df['diagnosis']
```

```python
data_dia = y
data = x

data_n_2 = (data - data.mean()) / (data.std()) # 자체 표준화
data = pd.concat([y, data_n_2.iloc[:, 0:10]], axis=1) # 변수가 너무 많아서 10개씩 끊어줌
data = pd.melt(data, id_vars="diagnosis", var_name="features", value_name="value") # column을 type처럼 변수로 바꿔서 데이터를 재구조화해주는 역할인듯

plt.figure(figsize=(10, 10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=data, split=True, inner='quart')
plt.xticks(rotation=90)
```

```python
data = pd.concat([y, data_n_2.iloc[:, 10:20]], axis=1)
data = pd.melt(data, id_vars="diagnosis", var_name="features", value_name='value')

plt.figure(figsize=(10, 10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=data, split=True, inner='quart')
plt.xticks(rotation=90)
```

```python
data = pd.concat([y, data_n_2.iloc[:, 20:]], axis=1)
data = pd.melt(data, id_vars="diagnosis", var_name="features", value_name="value")

plt.figure(figsize=(10, 10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=data, split=True, inner='quart')
plt.xticks(rotation=90)
```

```python
# boxplot의 경우, 이상치 확인하기도 좋은 도구

plt.figure(figsize=(10, 10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=data)
plt.xticks(rotation=-45)
```

```python
sns.jointplot(x.loc[:, 'concavity_worst'], x.loc[:, 'concave points_worst'], kind="reg")
# 두 변수간 상당히 강한 상관관계를 가진 것으로 보임
```

```python
sns.set(style='white')

# 데이터 프레임에서 특정 열 선택
df = x.loc[:, ['radius_worst', 'perimeter_worst', 'area_worst']]

# PairGrid 생성
g = sns.PairGrid(df, diag_sharey=False)
# 하삼각 영역에 kdeplot 사용
g.map_lower(sns.kdeplot, cmap="Blues_d")
# 상삼각 영역에 산점도 적용
g.map_upper(plt.scatter)
# 대각선 영역에 커널 밀도 추정 사용
g.map_diag(sns.kdeplot, lw=3)
```

```python
sns.set(style="whitegrid", palette='muted')

data_dia = y
data = x
data_n_2 = (data - data.mean()) / (data.std())
data = pd.concat([y, data_n_2.iloc[:, 0:10]], axis=1)
data = pd.melt(data, id_vars="diagnosis", var_name="features", value_name='value')

plt.figure(figsize=(10, 10))

import time
tic = time.time()
sns.swarmplot(x="features", y="value", hue="diagnosis", data=data)

plt.xticks(rotation=-45)
```

```python
# 큰 상관이 없는 column들 삭제
drop_list = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x_1 = x.drop(columns=drop_list)
x_1.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, confusion_matrix, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x_1, y, test_size=0.3, random_state=42)

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

pred = rf.predict(x_test)
ac = accuracy_score(y_test, pred)
print(ac)

cm = confusion_matrix(y_test, pred)
sns.heatmap(cm, fmt="d")
```
